### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,11)
y=x**2
plt.plot(x,y,marker='o',color='blue',
         linestyle='-')
plt.title('Line plot of y=x2')
plt.xlabel('x values')
plt.ylabel('y values (x squared)')
plt.grid(True)
plt.show()
```
