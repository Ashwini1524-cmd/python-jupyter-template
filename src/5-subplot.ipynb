### Exercise 5: Subplot Layout

```python
#
import matplotlib.pyplot as plt

# Create figure and 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line Chart (Top-left)
x = [0, 1, 2, 3, 4, 5]
y = [i**2 for i in x]
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Chart')

# Bar Chart (Top-right)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Chart')

# Scatter Plot (Bottom-left)
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter)
axs[1, 0].set_title('Scatter Plot')

# Pie Chart (Bottom-right)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Layout adjustment
plt.tight_layout()
plt.show()
```
