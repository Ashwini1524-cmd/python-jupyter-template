### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt
categories=['A','B','C','D']
values=[5,7,3,9]
colors=['red','green','blue','orange']
plt.bar(categories,values,color=colors)
plt.title('Bar Plot of category Values')
plt.xlabel('Categories')
plt.ylabel('values')
plt.grid(axis='y')
plt.show()
```
